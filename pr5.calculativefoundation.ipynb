## PART A: VECTOR & MATRIX FUNDAMENTALS

```python
import numpy as np
import pandas as pd
```

```python
# Rows = Students, Columns = Subjects (Math, Science, English, Computer)
data = np.array([
    [85, 78, 92, 88],
    [75, 82, 79, 85],
    [90, 88, 94, 92],
    [60, 65, 70, 68],
    [80, 78, 85, 82]
])

students = ["S1", "S2", "S3", "S4", "S5"]
subjects = ["Math", "Science", "English", "Computer"]

df = pd.DataFrame(data, index=students, columns=subjects)
df
```

```python
#1. Represent each student as vector

v1 = data[0]  # Student 1
v2 = data[1]  # Student 2

print("Student 1 Vector:", v1)
print("Student 2 Vector:", v2)
```

```python
#2. Calculate
#a) Norm-1 and Norm-2

# L1 Norm
l1 = np.linalg.norm(v1, ord=1)

# L2 Norm
l2 = np.linalg.norm(v1)

print("L1 Norm:", l1)
print("L2 Norm:", l2)
```

```python
#b) Dot Product & Angle

dot = np.dot(v1, v2)

# angle
angle = np.arccos(dot / (np.linalg.norm(v1) * np.linalg.norm(v2)))
angle_deg = np.degrees(angle)

print("Dot Product:", dot)
print("Angle (degrees):", angle_deg)
```

```python
# c)Cross Product

cross = np.cross(v1[:3], v2[:3])
print("Cross Product:", cross)
```

```python
# 3) Projection of one vector onto another

projection = (np.dot(v1, v2) / np.dot(v2, v2)) * v2
print("Projection of v1 on v2:", projection)
```

## PART B: MATRIX OPERATIONS

```python
#4) Matrix Representation

A = data
print("Matrix A:\n", A)
```

```python
#a) Matrix Addition & Multiplication

# Addition
add = A + A

# Multiplication
mul = np.dot(A, A.T)

print("Addition:\n", add)
print("\nMultiplication:\n", mul)
```

```python
#b) Transpose, Inverse, Determinant

# Transpose
transpose = A.T

# Square matrix for inverse
B = A[:4, :4]

det = np.linalg.det(B)
inv = np.linalg.inv(B)

print("Transpose:\n", transpose)
print("\nDeterminant:", det)
print("\nInverse:\n", inv)
```

## PART C: LINEAR TRANSFORMATION

```python
# 5) Line, Plane, Hyperplane

'''
Line (1D) → vector span
Plane (2D) → combination of 2 vectors
Hyperplane → higher dimensions'''


# Linear combination
a, b = 2, 3
result = a*v1 + b*v2
print(result)
```

```python
# 6) Dimensionality Increase

# Original data (4D)
print("Original Shape:", A.shape)

# Add new feature: Total marks (new dimension)
total = np.sum(A, axis=1).reshape(-1, 1)

A_new = np.hstack((A, total))

print("New Data:\n", A_new)
print("New Shape:", A_new.shape)
```

## PART D: EIGENVALUES & DECOMPOSITION

```python
#7) Covariance Matrix + Eigenvalues

# Covariance matrix
cov = np.cov(A, rowvar=False)

# Eigenvalues & vectors
eigenvalues, eigenvectors = np.linalg.eig(cov)

print("Covariance Matrix:\n", cov)
print("\nEigenvalues:\n", eigenvalues)
print("\nEigenvectors:\n", eigenvectors)
```

```python
#8) LU Decomposition

from scipy.linalg import lu

P, L, U = lu(B)

print("L:\n", L)
print("\nU:\n", U)
```

```python
#9) Singular Value Decomposition (SVD)

U, S, Vt = np.linalg.svd(A)

print("U:\n", U)
print("\nSingular Values:\n", S)
print("\nVt:\n", Vt)
```

## PART E: DIMENSIONALITY REDUCTION

```python
#10) PCA (Reduce to 2D)

from sklearn.decomposition import PCA

pca = PCA(n_components=2)
reduced = pca.fit_transform(A)

print("Reduced Data (2D):\n", reduced)
```

```python
#11) LDA Classification

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

# Labels: Above avg (1), Below avg (0)
labels = [1, 1, 1, 0, 1]

lda = LinearDiscriminantAnalysis(n_components=1)
lda_result = lda.fit_transform(A, labels)

print("LDA Result:\n", lda_result)
```
